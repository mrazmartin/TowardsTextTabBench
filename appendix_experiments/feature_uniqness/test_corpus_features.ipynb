Here is just a short study over how many unique values features of some CARTE datasets have. We are particullarly interested in the textual ones. 

This can be extended into a further study about how long these textual features are, even when they show some level of uniqueness... 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read and clean header
with open("corpus.csv", "r", encoding="utf-8") as f:
    # Read first two lines which together form the real header
    header_line_1 = f.readline().strip().split(",")
    header_line_2 = f.readline().strip().split(",")

# Merge multi-line headers
final_headers = [
    h1.strip('"') + " " + h2.strip('"') if h2 else h1.strip('"')
    for h1, h2 in zip(header_line_1, header_line_2 + [""] * (len(header_line_1) - len(header_line_2)))
]

# Now read the rest of the CSV with the corrected headers
df = pd.read_csv("corpus.csv", skiprows=2, names=final_headers)

# Compute number of unique values per column
unique_counts = df.nunique()

# Plot the results
plt.figure(figsize=(12, 6))
unique_counts.sort_values(ascending=False).plot(kind='bar')
plt.title('Number of Unique Values per Column')
plt.ylabel('Unique Count')
plt.xlabel('Column Name')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("drugs.csv")

# Drop completely empty columns
df = df.dropna(axis=1, how='all')

# Drop columns with names starting with 'Unnamed' or with <50% non-null values
threshold = len(df) * 0.5
df = df.loc[:, ~df.columns.str.startswith("Unnamed")]
df = df.dropna(axis=1, thresh=threshold)

# Compute number of unique values per column
unique_counts = df.nunique()
sorted_counts = unique_counts.sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 6))
ax = sorted_counts.plot(kind='bar')
plt.title('Number of Unique Values per Column (Cleaned Drugs Dataset)')
plt.ylabel('Unique Count')
plt.xlabel('Column Name')
plt.xticks(rotation=90)

# Add number labels above bars
for i, v in enumerate(sorted_counts):
    ax.text(i, v + 0.2, str(v), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
